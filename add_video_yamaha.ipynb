```python
import pandas as pd
import numpy as np
import re
```

```python
input_file = 'yamaha_trumpets.csv'
output_file = f'{input_file[:-4]}_output.csv'
```

```python
df = pd.read_csv(input_file)
df.head()
```

```python
video_tag = """<iframe width="560" height="315" src="https://www.youtube.com/embed/Da71wAwb_8Y" title="YouTube video player" frameborder="0" allow="accelerometer; autoplay; clipboard-write; encrypted-media; gyroscope; picture-in-picture; web-share" allowfullscreen></iframe>
"""
```

```python
df['Body (HTML)'] = df['Body (HTML)'] + video_tag
```

```python
try:
    df = df.replace(np.nan, '')
    df['Variant Barcode'] = df['Variant Barcode'].astype(str).apply(lambda x: re.sub( r'\.0$', '', x) )
    df['Variant SKU'] = df['Variant SKU'].astype(str).apply(lambda x: re.sub( r'\.0$', '', x) )
    df.to_csv(output_file, index=False)
    print(f'Created ouput file {output_file}')
except:
    raise RuntimeError("Couldn't save csv file")
```
