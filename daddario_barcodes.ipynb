```python
# init

import pandas as pd
import re
```

```python
input_file = "rico_products_lm_updated.csv"
output_file = "rico_products_lm_barcodes_updated.csv"

df = pd.read_csv(input_file)
```

```python
df.head()
```

```python
df['Variant Barcode'] = '0' + df['Variant Barcode'].astype(str)
```

```python
df['Variant Barcode'] = df['Variant Barcode'].astype(str).apply(lambda x: re.sub( r'\.0$', '', x) )
```

```python
df['Variant Barcode'] = df['Variant Barcode'].str.replace('0nan', '')
```

```python
df['Variant Barcode']
```

```python
df.to_csv(output_file, index=False)
```
